```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)

x1 = [1,2,3]
x2 = np.array([[1,4,7],[2,5,8],[3,6,9]])

print(softmax(x1))
print(softmax(x2))
```

```python
x = np.arange(-1.0, 3.0, 1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])
print(x)
print(scores)
print(softmax(scores))
#scores = x

plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```

```python
def sigmoid(x):
    x = 1 / (1 + np.exp(-x))
    return x / sum(x)

print(x)
print(scores)
print(sigmoid(scores))
print(np.sum(sigmoid(scores), axis=0))

plt.plot(x, sigmoid(scores).T, linewidth=2)
plt.show()
```

```python
def wtf(x):
    x = np.square(x)
    return x / np.sum(x, axis=0)

print(x)
print(scores)
plt.plot(x, scores.T, linewidth=2)
plt.show()
print(wtf(scores))

plt.plot(x, wtf(scores).T, linewidth=2)
plt.show()
```
